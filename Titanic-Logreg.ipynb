```python
x = df.drop('Survived',axis=1)
y = df['Survived']
```

```python
# scaler = MinMaxScaler(feature_range=(0,1))
scaler = StandardScaler()
# scaler = Normalizer()
x_scaled = scaler.fit_transform(x)
```

```python
logreg = LogisticRegression(C=2)
```

```python
x_train,x_valid,y_train,y_valid = train_test_split(x_scaled,y)
logreg.fit(x_train,y_train)
print(logreg.score(x_valid,y_valid))
scores = cross_val_score(logreg,x_scaled,y,cv = 10,scoring="accuracy")
print(scores)
print(scores.mean())
```

```python
c_range_small = list(np.arange(0.1,1,0.1))
c_range_large = list(range(1,100))
c_range = c_range_small + c_range_large
scoreslist=[]
for cval in c_range:
    logreg = LogisticRegression(C=cval)
    scores = cross_val_score(logreg,x_scaled,y,cv = 10,scoring="accuracy")
    #print("score for c:",cval," =",scores.mean())
    scoreslist.append(scores.mean())
plt.figure(figsize=(20,3)) 
plt.plot(c_range[:20],scoreslist[:20])
```
